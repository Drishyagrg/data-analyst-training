# if-elif statement
syntax:
if condition:
    # code
elif condition:
    # code
    ......
else:
    # code

```python
a=int(input("enter number 1"))
b=int(input("enter number 2"))
c=int(input("enter number 3"))
if a>b and a>c:
    print(f'{a} is greater number')
elif b>a and b>c:
    print(f'{b} is greater number')
elif c>a and c>b:
    print(f'{c} is greater number')
else:
    print('All the numbers are equal')
```

```python
# enter a number from the user and check if it is divisible by 3,5 and both or none
n=int(input("Enter a number"))
if n%3==0:
    print(f'{n} is divisible by 3')
elif n%5==0:
    print(f'{n} is divisible by 5')
elif n%3==0 and n%5==0:
    print(f'{n} is divisible by both 3 and 5')
else:
    print(f'{n} is not divisible by both 3 and 5')
```

# nested if
- if statement defined inside another if statement
syntax:

```python
# find whether the number is even or odd if the number is +ve.
num= int(input("Enter a number"))
if num>0:
    if num%2==0:
        print(f'{num} is even')
    else:
        print(f'{num} is odd')
else: 
    print("The number is negative")
```

```python
# Enter a number from the user and check whether it is multiple of 3 
# if it is multiple of 3 then check if it is divisible by 7.
num=int(input("Enter a number"))
if num%3==0:
    if num%7==0:
        print(f'{num} is multiple of 3 and is divisible by 7')
    else:
        print(f'{num} is multiple of 3 but is not divisible by 7')
else:
    print(f'{num} is not multiple of 3')
```

```python
a=[1,2,3,4,5,6,7,8,9]
odd=[]
even=[]
for i in a:
    # print(i)
    if i%2==0:
        print('even ', i)
        even.append(i)
    else:
        print('odd ', i)
        odd.append(i)
```

```python
print('odd =',odd)
print('even =',even)
```

```python
# make a collection of string and seperate those are even or not
s1=['nim','pradeepti','Nimbuti','pradeep','drishh','drish','nimmmmmm']
evens1=[]
odds1=[]
for i in s1:
    if len(i)%2==0:
        print('even string ', i)
        evens1.append(i)
    else:
        print('odd string ',i)
        odds1.append(i)
```
