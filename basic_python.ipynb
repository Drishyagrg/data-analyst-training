```python
# Check whether the numbers are Armstrong or not
a=[121,153,370,570]
armstrong_num=[]
notarmstrong_num=[]
for i in a:
    length=len(str(i))
    sum=0
    temp=i
    while temp>0:
        rem= temp%10
        sum=sum +rem**length;
        temp//=10
    if sum== i:
        print(f'{i} is an armstrong number')
        armstrong_num.append(i)
        
    else:
        print(f'{i} is not an armstrong number')
        notarmstrong_num.append(i)
    
```

```python
print('Armstrong Collection', armstrong_num)
print('Non Armstrong collection', notarmstrong_num)
```

```python
# odd or even
a=[12,34,2,4645,675,3,56,34,46]
def even_or_odd(num):
    even=[]
    odd=[]
    for a in num:
        if a%2==0:
            even.append(a)
        else:
            odd.append(a)
    return even, odd

even, odd = even_or_odd(a)
print('even number is ',even)
print('odd number is ',odd )
```

```python
# find factorial of a number
def factorial_num(a):
    n=a
    fact=1
    while a>0:
        fact=fact*a
        a=a-1
    print(f'factorial of {n} is {fact}')

factorial_num(5)
```

```python
# 1. check the number is prime or not?
def prime(num):
    temp=num-1
    count=0
    while temp>1:
        # print(temp)
        if num%temp==0:
            count+=1
        temp=temp-1
    if count==0:
        print(f'{num} is a prime number')
    else:
        print(f'{num} is not a prime number')
    print(count)

prime(2)
            
```

```python
# 2. check in the collection whether a number are perfect number or not?
def perfect_num(num1):
    sum=0
    temp=num1-1
    while temp>0:
        if num1%temp==0:
            sum+=temp
        temp-=1
    if sum==num1:
        print(f'{num1} is a perfect number')
    else:
        print(f'{num1} is not a perfect number')

perfect_num(60)
        
```

```python
# 4. check a number is +ve or not, if +ve check it is divisible by 11 or not?
def positive(num3):
    if num3>0:
        if num3%11==0:
            print(f'{num3} is a +ve number and divisible by 11')
        else:
            print(f'{num3} is a +ve number but not divisible by 11')
    else:
        print(f'{num3} is a -ve number')

positive(-55)
```

```python
# 5. check square of a number is prime or not?
def square(num4):
    num4= num4**2
    # print(num4)
    temp=num4-1
    count=0
    while temp>1:
        # print(temp)
        if num4%temp==0:
            count+=1
        temp=temp-1
    if count==0:
        print(f'{num4} is a prime number')
    else:
        print(f'{num4} is not a prime number')
    print(count)

square(2)
    
```

```python
# 6. in a collection check a number is perfect number or not?
numbers=[2,4534,75646,245,6,28,58,56]
perfect_num=[]
non_perfect_num=[]
def perfect_num_collection(numbers):
    for i in numbers:
        sum=0
        temp=i-1
        while temp>0:
            if i%temp==0:
                sum+=temp
            temp-=1
        if sum==i:
            print(f'{i} is a perfect number')
            perfect_num.append(i)
        else:
            print(f'{i} is not a perfect number')
            non_perfect_num.append(i)
    print('perfect_num',perfect_num)
    print('non_perfect_num', non_perfect_num)

perfect_num_collection(numbers)

```
